```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Load the dataset
df = pd.read_csv('Advertising.csv')

# Simulate a date-based index for visualization
df.index = pd.date_range(start='2024-01-01', periods=len(df), freq='D')

# Plot raw sales data
df.plot(y='Sales', figsize=(16, 6), fontsize=12)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.title('Sales Over Simulated Time')
plt.grid(True)
plt.show()

# Calculate rolling average (e.g., 7-day window)
df['Sales_Rolling'] = df['Sales'].rolling(window=7, center=True).mean()

# Plot smoothed sales
df[['Sales_Rolling']].plot(figsize=(16, 6), linewidth=2)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Smoothed Sales', fontsize=12)
plt.title('Smoothed Sales Trend (7-Day Rolling Avg)')
plt.grid(True)
plt.legend(['Smoothed Sales'])
plt.show()
```
