```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pumpkins = pd.read_csv(r'C:\Users\Lacuesta\Downloads\US-pumpkins.csv')
pumpkins['Low Price'] = pd.to_numeric(pumpkins['Low Price'], errors='coerce')
pumpkins['High Price'] = pd.to_numeric(pumpkins['High Price'], errors='coerce')
pumpkins['Price'] = (pumpkins['Low Price'] + pumpkins['High Price']) / 2
pumpkins['Month'] = pd.to_datetime(pumpkins['Date'], format='%m/%d/%y').dt.month
new_pumpkins = pumpkins.dropna(subset=['Price', 'Month'])
```

```python
plt.scatter(new_pumpkins['Price'], new_pumpkins['Month'])
plt.xlabel("Price")
plt.ylabel("Month")
plt.title("Pumpkin Price vs. Month")
plt.show()
```

```python
new_pumpkins.groupby('Month')['Price'].mean().plot(kind='bar')
plt.ylabel("Average Pumpkin Price")
plt.xlabel("Month")
plt.title("Average Pumpkin Price by Month")
plt.show()
```
