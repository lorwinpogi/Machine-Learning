```python
import matplotlib.pyplot as plt
import seaborn as sns


numeric_cols = [
    'release_date', 'length', 'popularity', 'danceability', 'acousticness',
    'energy', 'instrumentalness', 'liveness', 'loudness',
    'speechiness', 'tempo', 'time_signature'
]

corrmat = df[numeric_cols].corr()
df = df[df['artist_top_genre'] != 'Missing']
top = df['artist_top_genre'].value_counts()
plt.figure(figsize=(10,7))
sns.barplot(x=top.index,y=top.values)
plt.xticks(rotation=45)
plt.title('Top genres',color = 'blue')
```

```python
# Plot heatmap
plt.figure(figsize=(12, 9))
sns.heatmap(corrmat, vmax=0.8, square=True, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Nigerian Songs", fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
