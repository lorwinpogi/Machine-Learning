```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("nigerian-songs.csv")

sns.set(style="whitegrid")
```

```python
# Create subplots for each feature
plt.figure(figsize=(20, 20), dpi=200)


# List of features to plot
features = [
    'popularity', 'acousticness', 'energy', 'instrumentalness',
    'liveness', 'loudness', 'speechiness', 'tempo',
    'time_signature', 'danceability', 'length', 'release_date'
]
# Loop through features and create subplots
for i, feature in enumerate(features, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(x=feature, data=df)
    plt.title(f'Boxplot of {feature}')
    plt.xlabel('')  # Remove x-axis label for cleaner visuals

plt.tight_layout()
plt.show()
```
