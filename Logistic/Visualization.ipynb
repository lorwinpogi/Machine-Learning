```python
import pandas as pd

# Use correct filename and check it's in your working directory
df = pd.read_csv('US-pumpkins.csv')

# View actual column names to avoid typos
print(df.columns)
```

```python
required_columns = ['City Name', 'Package', 'Variety', 'Origin', 'Item Size', 'Color']
for col in required_columns:
    if col not in df.columns:
        print(f"Missing column: {col}")
```

```python
columns_to_select = ['City Name', 'Package', 'Variety', 'Origin', 'Item Size', 'Color']
pumpkins = df.loc[:, columns_to_select].dropna()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

palette = {
    'ORANGE': 'orange',
    'WHITE': 'wheat',
}

sns.catplot(
    data=pumpkins,
    y="Variety",
    hue="Color",
    kind="count",
    palette=palette
)

plt.show()
```
